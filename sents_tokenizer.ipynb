```python
import os
import re
import pandas as pd
from nltk import sent_tokenize
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
from sklearn.metrics import classification_report
```

```python
with open("text.txt", encoding="cp1251") as file:
    data = file.read()
    data = re.sub('\n+', ' ', data)
    data = data.replace('\xa0', ' ')
```

```python
data
```

```python
tokenized_sents = sent_tokenize(data)
```

```python
tokenized_sents
```

```python
#полученное я затем исправила руками, записала в файл sents_tokenized.txt и оттуда прочитала
with open("sents_tokenized.txt") as file:
    tokenized_sents = [re.sub('[\n\t]', '', x) for x in file.readlines()]
```

```python
tokenized_sents
```

```python
regex = '[.!?]+ +[А-Я]+'

sents_tokenized_by_regex = re.split(regex, data)
```

```python
sents_tokenized_by_regex
```

Для подсчёта качества используем это: "Есть ещё такая метрика - берем идеальное предложение и пробуем разбить его нашей моделью - если  разбивается, считает этой ошибкой, если нет - правильным ответом. 
Потом берем два идеальных предложения - склеиваем их пробельным символом и пробуем разбить нашей моделью. Если разбивает на два нужных предложения - все правильно. Таким образом мы вычисляем tp fp fn для расчета точности, полноты и f1 меры."

```python
tp = 0
fp = 0
fn = 0

for sent in tokenized_sents:
    if len(re.split(regex, sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(tokenized_sents)-1):
    sent1, sent2 = tokenized_sents[i], tokenized_sents[i+1]
    sent = ' '.join([sent1, sent2])
    if len(re.split(regex, sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision) #false positives вообще нет
print('Recall - ', recall)
print('F1 - ', f1)
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train(data)
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
sents_tokenized_by_punkt = tokenizer.tokenize(data)
```
