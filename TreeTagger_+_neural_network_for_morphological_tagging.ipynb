```
import numpy as np
from lxml import etree
from collections import defaultdict
from collections import Counter
```

```
open_corpora = etree.fromstring(open('annot.opcorpora.no_ambig_strict.xml', 'rb').read())
```

```
from sklearn.model_selection import KFold
```

```
#делаем кросс-валидацию – 5 раз случайно разбиваем наш корпус на трейн и тест, результаты записываем в corpus_train_n.txt и corpus_test_n.txt

vocab = defaultdict(set)
tags = set()

kf = KFold(n_splits=5)
splits = list(kf.split(open_corpora.xpath('//tokens')))
for i in range(len(splits)):
    corpus_train = open(f'corpus_train_{i}.txt', 'w')
    corpus_test = open(f'corpus_test_{i}.txt', 'w')
    corpus_test_tags = open(f'corpus_test_tags_{i}.txt', 'w')
    #для тестовых предложений будем записывать сами предложения в один файл, а тэги – в другой
    train_index, test_index = splits[i][0], splits[i][1]
    sents_train = np.array(open_corpora.xpath('//tokens'))[train_index]
    sents_test = np.array(open_corpora.xpath('//tokens'))[test_index]
    for sent in sents_train:
        length = len(sent.xpath('token'))
        ended = False
        for i,token in enumerate(sent.xpath('token')):
            word = token.xpath('@text')
            gram_info = token.xpath('tfr/v/l/g/@v')

            if (i+1)==length and gram_info[0] == 'PNCT':
                gram_info = ['SENT']
                ended = True

            corpus_train.write(word[0] + '\t' + ','.join(gram_info) + '\n')
            lemma = token.xpath('tfr/v/l/@t')[0]
            vocab[word[0].lower()].add((','.join(gram_info), lemma.lower()))
            tags.add(','.join(gram_info))

        if not ended:
            corpus_train.write('.\tSENT\n')
    for sent in sents_test:
        length = len(sent.xpath('token'))
        ended = False
        for i,token in enumerate(sent.xpath('token')):
            word = token.xpath('@text')
            gram_info = token.xpath('tfr/v/l/g/@v')

            if (i+1)==length and gram_info[0] == 'PNCT':
                gram_info = ['SENT']
                ended = True

            corpus_test.write(word[0] + '\n')
            corpus_test_tags.write(','.join(gram_info) + '\n')
            lemma = token.xpath('tfr/v/l/@t')[0]
            vocab[word[0].lower()].add((','.join(gram_info), lemma.lower()))
            tags.add(','.join(gram_info))

        if not ended:
            corpus_test.write('SENT\n')
            corpus_test_tags.write('SENT\n')
    corpus_train.close()
    corpus_test.close()
    corpus_test_tags.close()
```

Прежде, чем обучать, посчитаем меру Жаккара для pymorphy, чтобы нам было с чем сравнивать качество нашего тэггера.

```
from pymorphy2 import MorphAnalyzer

morph = MorphAnalyzer()

corpus = []

for sentence in open_corpora.xpath('//tokens'):
    sent_tagged = []
    for token in sentence.xpath('token'):
        word = token.xpath('@text')
        gram_info = token.xpath('tfr/v/l/g/@v')
        sent_tagged.append((word[0], set(gram_info)))
    
    corpus.append(sent_tagged)

preds_pymorphy = []
mistakes_pymorphy = Counter()

for sent in corpus:
    for word, tag in sent:
        pred = set(str(morph.parse(word)[0].tag).replace(' ', ',').split(','))
        p = len(pred&tag)/len(pred|tag)
        preds_pymorphy.append(p)
        if p < 0.5:
            mistakes_pymorphy.update([(word, tuple(tag), tuple(pred))])
```

```
means = []
mistakes = [Counter() for i in range(5)] #будем здесь хранить самые частотные ошибки на каждом фолде

for i in range(5):
    preds = []
    with open(f'corpus_test_{i}.txt') as file:
        words = (file.read()).split('\n')
    with open(f'corpus_test_tags_{i}.txt') as file:
        tags = (file.read()).split('\n')
    with open(f'output_{i}.txt') as file:
        output = (file.read()).split('\n')
    for j in range(len(words)):
        if tags[j] != 'SENT':
            word = words[j]
            tag = set(tags[j].split(','))
            pred = set(output[j].split(','))
            p = len(pred&tag)/len(pred|tag)
            preds.append(p)
            if p < 0.5:
                mistakes[i].update([(word, tuple(tag), tuple(pred))])
    means.append(np.mean(preds))
```

#Нейросеть

Обучаем нейросеть, которая предсказывает часть речи, одушевлённость, число, лицо, падеж и время.

```
import keras
from collections import Counter
```

Соберем корпус в список.

```
corpus = []
vocab = defaultdict(set)
tags = set()

for sentence in open_corpora.xpath('//tokens'):
    sent = []
    for token in sentence.xpath('token'):
        word = token.xpath('@text')
        gram_info = token.xpath('tfr/v/l/g/@v')  
        sent.append(word + gram_info)
    corpus.append(sent)

corpus_train = corpus[:8479] #это (приблизительно 0.8 всего корпуса) будем использовать для обучения
corpus_val = corpus[8479:] #а это – для валидации, когда будем считать меру жаккара
```

Создадим словарь для слов и частей речи. По одушевленности и другим тэгам проще построить словарь вручную.

```
vocab = Counter()
poses = Counter()
tags = Counter()

for sent in corpus_train:
    for word, pos, *tags in sent:
        vocab[word.lower()] += 1
        poses[pos] += 1
```

Возьмем только слова, встретившиеся 3+ раз.

```
vocab = {word for word,c in vocab.most_common() if c > 3}
```

```
len(vocab)
```

```
len(poses)
```

Сдалем словари - слово в индекс, тэг в индекс и наоборот.

UNK нужен для словарных слов.

```
id2word = {i+2:word for i, word in enumerate(vocab)}
id2word[0] = '<PAD>'
id2word[1] = '<UNK>'
word2id = {word:i for i, word in id2word.items()}
```

```
id2pos = {i+1:pos for i, pos in enumerate(poses)}
id2pos[0] = '<PAD>'
pos2id = {pos:i for i, pos in id2pos.items()}
```

```
id2anim = {1:'NONE', 2:'anim', 3:'inan'}
id2anim[0] = '<PAD>'
anim2id = {tag:i for i, tag in id2anim.items()}
```

```
id2num = {1:'NONE', 2:'sing', 3:'plur'}
id2num[0] = '<PAD>'
num2id = {tag:i for i, tag in id2num.items()}
```

```
id2per = {1:'NONE', 2:'1per', 3:'2per', 4:'3per'}
id2per[0] = '<PAD>'
per2id = {tag:i for i, tag in id2per.items()}
```

```
id2case = {1:'NONE', 2:'nomn', 3:'gent', 4:'datv', 5:'accs', 6:'ablt', 7:'loct', 8:'voct', 9:'gen2', 10:'acc2', 11:'loc2'}
id2case[0] = '<PAD>'
case2id = {tag:i for i, tag in id2case.items()}
```

```
id2tense = {1:'NONE', 2:'pres', 3:'past', 4:'futr'}
id2tense[0] = '<PAD>'
tense2id = {tag:i for i, tag in id2tense.items()}
```

Создадим 6 списков.

```
sents_ids = []
poses_ids = []
anim_ids = []
num_ids = []
per_ids = []
case_ids = []
tense_ids = []

for sent in corpus_train:
    sents_ids.append([word2id.get(word.lower(), 1) for word, *_ in sent])
    poses_ids.append([pos2id[tag] for word, tag, *_ in sent])
    anim_ids_sent = []
    num_ids_sent = []
    per_ids_sent = []
    case_ids_sent = []
    tense_ids_sent = []
    
    for word, pos, *tags in sent:
        anim_tag = 1
        num_tag = 1
        per_tag = 1
        case_tag = 1
        tense_tag = 1
        for tag in tags:
            if tag in anim2id:
                anim_tag = anim2id[tag]
            if tag in num2id:
                num_tag = num2id[tag]
            if tag in per2id:
                per_tag = per2id[tag]
            if tag in case2id:
                case_tag = case2id[tag]
            if tag in tense2id:
                tense_tag = tense2id[tag]
        anim_ids_sent.append(anim_tag)
        num_ids_sent.append(num_tag)
        per_ids_sent.append(per_tag)
        case_ids_sent.append(case_tag)
        tense_ids_sent.append(tense_tag)
    anim_ids.append(anim_ids_sent)
    num_ids.append(num_ids_sent)
    per_ids.append(per_ids_sent)
    case_ids.append(case_ids_sent)
    tense_ids.append(tense_ids_sent)
```

```
import numpy as np
np.mean([len(sent) for sent in sents_ids])
```

Приведем все предложения к одной длине. 10 больше средней длины, но и не очень большая, чтобы сильно усложнить модель.

```
MAX_LENGTH = 10
```

```
from keras.preprocessing.sequence import pad_sequences
 
sents_ids_padded = pad_sequences(sents_ids, maxlen=MAX_LENGTH, padding='post',  truncating='post')
poses_ids_padded = pad_sequences(poses_ids, maxlen=MAX_LENGTH, padding='post',  truncating='post')
anim_ids_padded = pad_sequences(anim_ids, maxlen=MAX_LENGTH, padding='post',  truncating='post')
num_ids_padded = pad_sequences(num_ids, maxlen=MAX_LENGTH, padding='post',  truncating='post')
per_ids_padded = pad_sequences(per_ids, maxlen=MAX_LENGTH, padding='post',  truncating='post')
case_ids_padded = pad_sequences(case_ids, maxlen=MAX_LENGTH, padding='post',  truncating='post')
tense_ids_padded = pad_sequences(tense_ids, maxlen=MAX_LENGTH, padding='post',  truncating='post')
```

Разбиваем выборку на трейн и тест.

```
from sklearn.model_selection import train_test_split
train_index, test_index = train_test_split(list(range(sents_ids_padded.shape[0])),test_size=0.2)
```

```
train_sents, test_sents = sents_ids_padded[train_index], sents_ids_padded[test_index]
train_pos, test_pos = poses_ids_padded[train_index], poses_ids_padded[test_index]
train_anim, test_anim = anim_ids_padded[train_index], anim_ids_padded[test_index]
train_num, test_num = num_ids_padded[train_index], num_ids_padded[test_index]
train_per, test_per = per_ids_padded[train_index], per_ids_padded[test_index]
train_case, test_case = case_ids_padded[train_index], case_ids_padded[test_index]
train_tense, test_tense = tense_ids_padded[train_index], tense_ids_padded[test_index]
```

Создаем модель (у модели 6 выходов).

```
from keras.models import Model
from keras.layers import Dense, LSTM, Input, Bidirectional, TimeDistributed, Embedding, Activation
from keras.optimizers import Adam
 

inp = Input(shape=(MAX_LENGTH, ))
# создаем эмбеддинги размерностью 8
x = Embedding(len(word2id), 8)(inp)

# пропускаем эмбеддинги через LSTM, чтобы модель учитывала контекст
x = LSTM(20, return_sequences=True)(x)

# каждый выход - последовательность классов
pos = TimeDistributed(Dense(len(pos2id), activation='softmax'))(x)
anim = TimeDistributed(Dense(len(anim2id), activation='softmax'))(x)
num = TimeDistributed(Dense(len(num2id), activation='softmax'))(x)
per = TimeDistributed(Dense(len(per2id), activation='softmax'))(x)
case = TimeDistributed(Dense(len(case2id), activation='softmax'))(x)
tense = TimeDistributed(Dense(len(tense2id), activation='softmax'))(x)

 
model= Model(inputs=inp, outputs=[pos, anim, num, per, case, tense], 
           )
 
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```
from keras.utils import to_categorical
```

Обучаем.

```
model.fit(x=train_sents, 
        y=[to_categorical(train_pos),
           to_categorical(train_anim),
           to_categorical(train_num),
           to_categorical(train_per),
           to_categorical(train_case),
           to_categorical(train_tense)], 
          batch_size=128, epochs=30, validation_split=0.2)
 #лосс стабильно снижается, метрики стабильно растут
```

Протестируем и посмотрим на метрики.

```
scores = model.evaluate(x=test_sents, y=[to_categorical(test_pos),
                                         to_categorical(test_anim),
                                         to_categorical(test_num),
                                         to_categorical(test_per),
                                         to_categorical(test_case),
                                         to_categorical(test_tense)])
```

```
#пишем свою функцию предикта, чтобы модель не предсказывала грамматические категории, не существующие у данных частей речи
#эта функция получает на вход массив предложений, уже разбитый по словам
#и возвращает массив формы кол-во предложений * кол-во слов * 6, где каждому слову в предложении соответствует массив его тэгов
def predict(model, sents):
    result = [[] for i in range(len(sents))]
    sents_X = []
    for sent in sents:
        sent_ids = []
        for word in sent:
            sent_ids.append(word2id.get(word.lower(), 1))
        sents_X.append(sent_ids)
    sents_X = pad_sequences(sents_X, maxlen=MAX_LENGTH, padding='post')
    predictions = model.predict(sents_X)
    for i in range(len(sents)):
        for j, idx in enumerate(predictions[0].argmax(axis=2)[i]):
            if j > (len(sents[i])-1):
                break
            result[i].append([id2pos[idx], 'NONE', 'NONE', 'NONE', 'NONE', 'NONE'])
        for j, idx in enumerate(predictions[1].argmax(axis=2)[i]):
            if j > (len(sents[i])-1):
                break
            if result[i][j][0] == 'NOUN':
                result[i][j][1] = id2anim[idx]
        for j, idx in enumerate(predictions[2].argmax(axis=2)[i]):
            if j > (len(sents[i])-1):
                break
            if result[i][j][0] in ['NOUN', 'ADJF', 'ADJS', 'VERB', 'PRTF', 'PRTS', 'NPRO']:
                result[i][j][2] = id2num[idx]
        for j, idx in enumerate(predictions[3].argmax(axis=2)[i]):
            if j > (len(sents[i])-1):
                break
            if result[i][j][0] in ['VERB', 'NPRO']:
                result[i][j][3] = id2per[idx]
        for j, idx in enumerate(predictions[4].argmax(axis=2)[i]):
            if j > (len(sents[i])-1):
                break
            if result[i][j][0] in ['NOUN', 'ADJF', 'PRTF', 'NPRO']:
                result[i][j][4] = id2case[idx]
        for j, idx in enumerate(predictions[5].argmax(axis=2)[i]):
            if j > (len(sents[i])-1):
                break
            if result[i][j][0] == 'VERB':
                result[i][j][5] = id2tense[idx]
    return result
```

Сделаем предсказания для отложенной выборки.




```
#так как нейросеть обучалась предсказывать не все граммемы, а только некоторые, при подсчёте меры Жаккара будем учитывать тэги только для этих граммем
used_tags = set(list(pos2id.keys())+
               list(anim2id.keys())+
               list(num2id.keys())+
               list(per2id.keys())+
               list(case2id.keys())+
               list(tense2id.keys()))

sents = [] #наши предложения
correct_tags = [] #здесь будем хранить "правильные" тэги

for sent in corpus_val:
    s = []
    tags = []
    for token in sent:
        s.append(token[0])
        tags.append(set(tag for tag in token[1:] if tag in used_tags))
    sents.append(s)
    correct_tags.append(tags)

predictions = predict(model, sents) #а здесь – то, что предсказала нейросеть
```

Посчитаем меру Жаккара между множествами правильных тэгов и результатом нейросети.

```
preds_neuro = []
mistakes_neuro = Counter()

for i in range(len(predictions)):
    for j in range(len(predictions[i])):
        word = sents[i][j]
        tags = correct_tags[i][j]
        pred = set(tag for tag in predictions[i][j] if tag != 'NONE')
        p = len(pred&tags)/len(pred|tags)
        preds_neuro.append(p)
        if p < 0.5:
            mistakes_neuro.update([(word, tuple(tags), tuple(pred))])
```
